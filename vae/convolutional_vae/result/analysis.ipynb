```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_loss = []
with open('mini_1_1000/train_loss.txt','r') as f:
    for line in f:
        train_loss.append(float(line.rstrip('\n')))
train_accuracy = []
with open('mini_1_1000/train_accuracy.txt','r') as f:
    for line in f:
        train_accuracy.append(float(line.rstrip('\n')))
```

```python
val_loss = []
with open('mini_1_1000/val_loss.txt','r') as f:
    for line in f:
        val_loss.append(float(line.rstrip('\n')))
val_accuracy = []
with open('mini_1_1000/val_accuracy.txt','r') as f:
    for line in f:
        val_accuracy.append(float(line.rstrip('\n')))
```

```python
print(len(val_accuracy),len(val_loss))
```

```python
x = np.arange(0, 1*len(train_accuracy), 1)
s = train_accuracy
t = val_accuracy

fig = plt.figure(figsize=(25,10))
fig.suptitle('Training and Validation accuracy on Mini-ImageNet dataset', fontsize=24)
plt.plot(x, t,'b')
plt.plot(x, s,'r')
plt.xlabel('iteration', fontsize=18)
plt.ylabel('accuracy', fontsize=18)
plt.gca().legend(('validation','training'), fontsize=18)
fig.savefig('accuracy.jpg')
plt.show()
```
