```python
def draw_plot(best1,best2,best3,best4,tick_name,le_name,save_name,save_op):

    flatui = ["#ee6000","#e7af00","#c16c02","#694001"]
    N = len(best1)
    fig = plt.figure(figsize=(16,9))
    ax=plt.subplot(1, 1, 1)
    ax.set_color_cycle(sns.color_palette(flatui))
    
    ind = np.arange(0,N*2,2)    # the x locations for the groups
    width = 0.35         # the width of the bars
    
    p1 = ax.bar(ind + 0*width, best1, width, bottom=0)
    p2 = ax.bar(ind + 1*width, best2, width, bottom=0)
    p3 = ax.bar(ind + 2*width, best3, width, bottom=0)
    p4 = ax.bar(ind + 3*width, best4, width, bottom=0)
    
    ax.set_xticks(ind + width)
    ax.set_xticklabels(tick_name,fontsize=12)
    ax.legend(le_name,fontsize=18,prop={'size': 12})
    plt.ylim((0.83,1.02))
    plt.ylabel('Testing Accuracy', fontsize=12)
    if save_op:
        fig.savefig(save_name,bbox_inches='tight')
    plt.show()
```

```python
best1 = [b_5151,b_5155,b_51201,b_51205]
best2 = [b_5551,b_5555,b_55201,b_55205]
best3 = [b_20151,b_20155,b_201201,b_201205]
best4 = [b_20551,b_20555,b_205201,b_205205]
tick_name = ['testing: 5-way 1-shot','testing: 5-way 5-shot',
             'testing: 20-way 1-shot','testing: 20-way 5-shot']
le_name = ['training: 5-way 1-shot','training: 5-way 5-shot','training: 20-way 1-shot','training: 20-way 5-shot']
save_name = './chinese_training.jpg'
sns.set(style="whitegrid")
draw_plot(best1,best2,best3,best4,tick_name,le_name,save_name,False)
```
