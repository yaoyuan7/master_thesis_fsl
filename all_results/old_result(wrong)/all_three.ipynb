```python
def draw_plot(best,proto,worst,tick_name,le_name,save_name,save_op):

    flatui = ["#53783b","#b3c656","#fcdb7e"]
    N = len(best)
    fig = plt.figure(figsize=(16,9))
    ax=plt.subplot(1, 1, 1)
    ax.set_color_cycle(sns.color_palette(flatui))
    
    ind = np.arange(0,N*2,2)    # the x locations for the groups
    width = 0.35         # the width of the bars
    
    p1 = ax.bar(ind + 0*width, best, width, bottom=0)
    p2 = ax.bar(ind + 1*width, worst, width, bottom=0)
    p3 = ax.bar(ind + 2*width, proto, width, bottom=0)

    ax.set_xticks(ind + width)
    ax.set_xticklabels(tick_name,fontsize=12)
    ax.legend(le_name,fontsize=18,prop={'size': 12})
    plt.ylim((0.85,1.02))
    plt.ylabel('Testing Accuracy', fontsize=12)
    if save_op:
        fig.savefig(save_name,bbox_inches='tight')
    plt.show()
```

```python
best = [b_51201,b_201201,b_601201,b_55201,b_205201,b_605201]
proto = [p_51201,p_201201,p_601201,p_55201,p_205201,p_605201]
worst = [w_51201,w_201201,w_601201,w_55201,w_205201,w_605201]
tick_name = ['Training: 5-way 1-shot \n Testing: 20-way 1-shot','Training: 20-way 1-shot \n Testing: 20-way 1-shot',
             'Training: 60-way 1-shot \n Testing: 20-way 1-shot','Training: 5-way 5-shot \n Testing: 20-way 1-shot',
             'Training: 20-way 5-shot \n Testing: 20-way 1-shot','Training: 60-way 5-shot \n Testing: 20-way 1-shot']
le_name = ['My model (multiple loss)','My model (single loss)','Prototypical network']
save_name = 'test-20way-1shot.jpg'
sns.set(style="whitegrid")
draw_plot(best,proto,worst,tick_name,le_name,save_name,True)
```
