```python
def draw_plot(best,proto,worst,tick_name,le_name,save_name,save_op):

    flatui = ["#53783b","#b3c656","#fcdb7e"]
    N = len(best)
    fig = plt.figure(figsize=(16,9))
    ax=plt.subplot(1, 1, 1)
    ax.set_color_cycle(sns.color_palette(flatui))
    
    ind = np.arange(0,N*2,2)    # the x locations for the groups
    width = 0.35         # the width of the bars
    
    p1 = ax.bar(ind + 0*width, best, width, bottom=0)
    p2 = ax.bar(ind + 1*width, worst, width, bottom=0)
    p3 = ax.bar(ind + 2*width, proto, width, bottom=0)

    ax.set_xticks(ind + width)
    ax.set_xticklabels(tick_name,fontsize=12)
    ax.legend(le_name,fontsize=18,prop={'size': 12})
    plt.ylim((0.94,1.01))
    plt.ylabel('Testing Accuracy', fontsize=12)
    if save_op:
        fig.savefig(save_name,bbox_inches='tight')
    plt.show()
```

```python
best = [b_51205,b_201205,b_601205,b_55205,b_205205,b_605205]
proto = [p_51205,p_201205,p_601205,p_55205,p_205205,p_605205]
worst = [w_51205,w_201205,w_601205,w_55205,w_205205,w_605205]
tick_name = ['Training: 5-way 1-shot \n Testing: 20-way 5-shot','Training: 20-way 1-shot \n Testing: 20-way 5-shot',
             'Training: 60-way 1-shot \n Testing: 20-way 5-shot','Training: 5-way 5-shot \n Testing: 20-way 5-shot',
             'Training: 20-way 5-shot \n Testing: 20-way 5-shot','Training: 60-way 5-shot \n Testing: 20-way 5-shot']
le_name = ['My model (multiple loss)','My model (single loss)','Prototypical network']
save_name = './prove/test-20way-5shot.jpg'
sns.set(style="whitegrid")
draw_plot(best,proto,worst,tick_name,le_name,save_name,True)
```

```python
def draw_two_plot(best_shot1,proto_shot1,best_shot5,proto_shot5,tick_name,le_name,save_name,save_op):

    flatui = ["#db5252","#ffbc78","#12339b","#7db6ff"]
    N = len(best_shot1)
    fig = plt.figure(figsize=(16,9))
    ax=plt.subplot(1, 1, 1)
    ax.set_color_cycle(sns.color_palette(flatui))
    
    ind = np.arange(0,N*2,2)    # the x locations for the groups
    width = 0.35         # the width of the bars
    
    p1 = ax.bar(ind + 0*width, best_shot1, width, bottom=0)
    p2 = ax.bar(ind + 1*width, proto_shot1, width, bottom=0)
    p3 = ax.bar(ind + 2*width, best_shot5, width, bottom=0)
    p4 = ax.bar(ind + 3*width, proto_shot5, width, bottom=0)

    ax.set_xticks(ind + 1.5*width)
    ax.set_xticklabels(tick_name,fontsize=12)
    ax.legend(le_name,fontsize=18,prop={'size': 12})
    plt.ylim((0.85,1.02))
    plt.ylabel('Testing Accuracy', fontsize=12)
    if save_op:
        fig.savefig(save_name,bbox_inches='tight')
    plt.show()
    
best_shot1 = [b_5151,b_20151,b_60151]
proto_shot1 = [p_5151,p_20151,p_60151]
best_shot5 = [b_5551,b_20551,b_60551]
proto_shot5 = [p_5551,p_20551,p_60551]
tick_name = ['Training: 5-way \n Testing: 5-way 1-shot', 'Training: 20-way \n Testing: 5-way 1-shot', 
             'Training: 60-way \n Testing: 5-way 1-shot']
le_name = ['My model (training: 1-shot)','Prototypical network (training: 1-shot)', 
           'My model (training: 5-shot)','Prototypical network (training: 5-shot)']
save_name = 'test-5way-1shot.jpg'
sns.set(style="whitegrid")
draw_two_plot(best_shot1,proto_shot1,best_shot5,proto_shot5,tick_name,le_name,save_name,True)
```
