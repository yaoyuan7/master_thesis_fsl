```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import t
import seaborn as sns
```

```python
_5w_1s_mean, _5w_1s_err = 0.9064, 0.0083
_10w_1s_mean, _10w_1s_err = 0.9132, 0.0081
_15w_1s_mean, _15w_1s_err = 0.9150, 0.0083
_20w_1s_mean, _20w_1s_err = 0.9248, 0.0074

_5w_5s_mean, _5w_5s_err = 0.9614, 0.0042
_10w_5s_mean, _10w_5s_err = 0.9665, 0.0039
_15w_5s_mean, _15w_5s_err = 0.9745, 0.0032
_20w_5s_mean, _20w_5s_err = 0.9747, 0.0033

_1shot_mean = [_5w_1s_mean,_10w_1s_mean,_15w_1s_mean,_20w_1s_mean]
_5shot_mean = [_5w_5s_mean,_10w_5s_mean,_15w_5s_mean,_20w_5s_mean]

_1shot_err = [_5w_1s_err,_10w_1s_err,_15w_1s_err,_20w_1s_err]
_5shot_err = [_5w_5s_err,_10w_5s_err,_15w_5s_err,_20w_5s_err]
```

```python
model = ['training 5-way','training 10-way','training 15-way','training 20-way']
x_pos = np.arange(0,4*5,5)
width = 1

sns.set(style="whitegrid")
fig = plt.figure(figsize=(15,8))
#flatui = ["#b1c0d4","#d4e5ea","#ffe7e7","#feffa8","#c4ffc1","#ffa5a5","#000000"]
flatui = ["#d4e5ea","#ffa5a5"]
ax=plt.subplot(1, 1, 1)
ax.set_color_cycle(sns.color_palette(flatui))

rects1 = ax.bar(x_pos-0.5, _1shot_mean, width,yerr=_1shot_err, 
                align='center', alpha=1, ecolor='black', capsize=6,label=r'testing: 5-way 1-shot')
rects2 = ax.bar(x_pos+0.5, _5shot_mean, width,yerr=_5shot_err, 
                align='center', alpha=1, ecolor='black', capsize=6,label=r'$testing: 5-way 5-shot')




ax.set_ylabel('Accuracy', fontsize=12)
ax.set_xticks(x_pos)
ax.set_ylim([0.8,1.1])
ax.set_xticklabels(model, fontsize=12)
#ax.set_title('Testing accuracy with 95% confident intervals')
ax.yaxis.grid(True)
ax.legend(fontsize=18,prop={'size': 12})

def autolabel(rects, xpos='center'):
    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0.1, 'right': 0, 'left': 1}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 1.05*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')

# Save the figure and show
plt.tight_layout()
#autolabel(rects1, "left")
#autolabel(rects2, "center")
#autolabel(rects3, "right")
#fig.savefig('error2.jpg',bbox_inches='tight')
plt.show()
```
